## *Machine Learning Survival Modeling*

### *Test: Coxnet model*

```python
# CoxPH elastic net model
coxnet = CoxPHElasticNetModel(train_df, test_df, batchNormType=batchNormType, dataName=dataName)
print(coxnet.hyperparameters)
coxnet_results = coxnet.train_over_subsets(
                           subset_sizes=[100,500],
                           runs_per_size=[20,20],
                           splits_per_size=[3,5],
                           trials_per_size=[20,20],
                           trial_threshold=20,
                           n_jobs=1,
                           is_tune=True,
                           is_save=False)
```

### *Test: SSVM*

```python
############## Some manual test ############
is_stratified = True
ssvm = SVMSurvivalModel(train_df, test_df, 
                        batchNormType=batchNormType, 
                        dataName=dataName, 
                        is_stratified=is_stratified)

## test out a single run
run = 0
time_col='time'
status_col='status'
batch_col="batch.id"


train_sub, _ = train_test_split(train_df, train_size=500, 
                                stratify=train_df[[status_col, batch_col]],
                                shuffle=True, random_state=run)
test_sub, _  = train_test_split(test_df,
                                train_size=1000,
                                stratify=test_df[[status_col, batch_col]],
                                shuffle=True, random_state=run)

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
batch_cols = pd.DataFrame(encoder.fit_transform(test_sub[[batch_col]]),
                        index=test_sub.index, 
                        columns=encoder.get_feature_names_out([batch_col]))
test_sub = pd.concat([batch_cols, test_sub.drop(columns=batch_col)], axis=1)

dataframe_to_scikitsurv_ds(test_sub)

### NOTE: implement naive approach of batch stratification by including batch.id as predictor
X_train, y_train = dataframe_to_scikitsurv_ds(train_sub)
X_test, y_test = dataframe_to_scikitsurv_ds(test_sub)

duration, tr_cind, te_cind, tr_brier, te_brier = ssvm.train(X_train, y_train, X_test, y_test)
                
print(f"(runtime: {duration:.2f}s)   |\
        (C-index)  Train: {tr_cind:.3f}, Test: {te_cind:.3f}   |\
        (Brier)  Train: {tr_brier:.3f}, Test: {te_brier:.3f}\n")  
```

```python
############ Sksurv pipeline #############
# Stratified SSVM

is_stratified = True
ssvm = SVMSurvivalModel(train_df, test_df, 
                        batchNormType=batchNormType, 
                        dataName=dataName, 
                        is_stratified=is_stratified)

ssvm_results = ssvm.train_over_subsets(
                           subset_sizes=[1000],
                           runs_per_size=[10],#20],
                           splits_per_size=[5],#5],
                           trials_per_size=[20],#20],
                           trial_threshold=20,
                           n_jobs=1,
                           is_tune=False,
                           is_save=True, 
                           fileName=None)
(
    ssvm_results
    .groupby('n train')
    .agg(mean=('test C', 'mean'),
        sd=('test C', 'std'))
    .reset_index()
)
```

```python
############ Sksurv pipeline #############
# non-stratified
is_stratified = False
ssvm_nonstrat = SVMSurvivalModel(train_df, test_df, 
                                batchNormType=batchNormType, 
                                dataName=dataName, 
                                is_stratified=is_stratified)

ssvm_results_nonstrat = ssvm_nonstrat.train_over_subsets(
                           subset_sizes=[1000],
                           runs_per_size=[10],#20],
                           splits_per_size=[5],#5],
                           trials_per_size=[20],#20],
                           trial_threshold=20,
                           n_jobs=1,
                           is_tune=False,
                           is_save=True, 
                           fileName=None)
(
    ssvm_results_nonstrat
    .groupby('n train')
    .agg(mean=('test C', 'mean'),
        sd=('test C', 'std'))
    .reset_index()
)
```

```python
modelname_dict = {'coxnet':'CoxPH',
                'svm': 'SSVM',
                'rsf': 'RSF',
                'deepsurv-torch': "DeepSurv"}

def load_simulate_results(dataFolderName, 
                        modelnames=['coxnet','svm','rsf','deepsurv-torch'],
                        fileName = 'model.results.10runs.txt'):
    
    results = pd.DataFrame({'n train': []})
    for mdl in modelnames:
        file_dir = os.path.join('models', dataFolderName, mdl, fileName)
        try:
            result_df = pd.read_table(file_dir, index_col=0)
        except:
            result_df = pd.read_table(os.path.join('models', dataFolderName, mdl, 'model.results.txt'), index_col=0)
        result_df['model'] = modelname_dict[mdl]
        results = pd.concat([results, result_df], axis=0)

    return results
```
